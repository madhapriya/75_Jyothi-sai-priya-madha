```python
import json
from io import StringIO
import pandas as pd
from pathlib import Path

DATA_DIR = Path("/content/sec_data")

def load_quarter(json_path):
    with open(json_path) as f:
        data = json.load(f)
    num = pd.read_csv(StringIO(data["num.txt"]), sep="\t")
    sub = pd.read_csv(StringIO(data["sub.txt"]), sep="\t")
    tag = pd.read_csv(StringIO(data["tag.txt"]), sep="\t")
    return num, sub, tag

num_all, sub_all, tag_all = [], [], []

for file in DATA_DIR.glob("20*q*.json"):
    n, s, t = load_quarter(file)
    num_all.append(n)
    sub_all.append(s)
    tag_all.append(t)

num_df = pd.concat(num_all, ignore_index=True)
sub_df = pd.concat(sub_all, ignore_index=True)
tag_df = pd.concat(tag_all, ignore_index=True).drop_duplicates("tag")
```

```python
KEY_TAGS = [
    "Revenues",
    "NetIncomeLoss",
    "OperatingIncomeLoss",
    "Assets",
    "Liabilities"
]

num_filt = num_df[num_df["tag"].isin(KEY_TAGS)].copy()
```

```python
df = (
    num_filt
    .merge(
        sub_df[["adsh", "name", "form", "period", "fy", "fp"]],
        on="adsh",
        how="left"
    )
    .merge(
        tag_df[["tag", "tlabel", "doc"]],
        on="tag",
        how="left"
    )
)

df["ddate"] = pd.to_datetime(df["ddate"], format="%Y%m%d", errors="coerce")
df.head()
```

```python
pivot = df.pivot_table(
    index="ddate",
    columns="tag",
    values="value",
    aggfunc="sum"
).sort_index()

pivot["Revenue_QoQ_%"] = pivot["Revenues"].pct_change(1) * 100
pivot["Revenue_YoY_%"] = pivot["Revenues"].pct_change(4) * 100
pivot["Operating_Margin_%"] = (
    pivot["OperatingIncomeLoss"] / pivot["Revenues"]
) * 100

kpi = pivot.dropna().round(2)
kpi.tail()
```

```python
import pandas as pd
import numpy as np
from io import StringIO
from pathlib import Path
```

```python
KEY_TAGS = [
    "Revenues",
    "NetIncomeLoss",
    "OperatingIncomeLoss",
    "Assets",
    "Liabilities"
]

num_filt = num_df[num_df["tag"].isin(KEY_TAGS)].copy()
```

```python
df = (
    num_filt
    .merge(
        sub_df[["adsh", "name", "form", "period", "fy", "fp"]],
        on="adsh",
        how="left"
    )
    .merge(
        tag_df[["tag", "tlabel", "doc"]],
        on="tag",
        how="left"
    )
)

df["ddate"] = pd.to_datetime(df["ddate"], format="%Y%m%d", errors="coerce")
df = df.sort_values(["name", "ddate"])
df.head()
```

```python
pivot = df.pivot_table(
    index="ddate",
    columns="tag",
    values="value",
    aggfunc="sum"
).sort_index()

pivot["Revenue_QoQ_%"] = pivot["Revenues"].pct_change(1) * 100
pivot["Revenue_YoY_%"] = pivot["Revenues"].pct_change(4) * 100
pivot["Operating_Margin_%"] = (
    pivot["OperatingIncomeLoss"] / pivot["Revenues"]
) * 100

kpi = pivot.dropna().round(2)
kpi.tail()
```

```python
from langchain_core.documents import Document

docs = []

for date, row in kpi.iterrows():
    text = f"""
    Period End: {date.date()}
    Revenue: {row['Revenues']}
    Revenue QoQ Change: {row['Revenue_QoQ_%']}%
    Revenue YoY Change: {row['Revenue_YoY_%']}%
    Operating Margin: {row['Operating_Margin_%']}%
    """

    docs.append(
        Document(
            page_content=text.strip(),
            metadata={"period": str(date.date())}
        )
    )

docs[0]

```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
```

```python
# Create local embeddings (NO API, NO QUOTA)
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Build vector database from your existing `docs`
vectordb = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory="./mdna_chroma"
)
```

```python
vectordb._collection.count()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_classic.chains.retrieval_qa.base import RetrievalQA
from langchain_core.prompts import PromptTemplate

PROMPT = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are a financial analyst drafting a Management’s Discussion and Analysis (MD&A).

RULES:
- Use ONLY the provided financial data
- Do NOT invent numbers
- Cite periods in square brackets like [2016-12-31]

Write the following sections:
1. Performance Trends
2. Revenue Drivers
3. Risks and Liquidity

Financial Data:
{context}

Task:
{question}

Tone: Formal, objective, SEC filing style.
"""
)

# ✅ ONLY MODEL THAT WORKS FOR YOU
llm = ChatGoogleGenerativeAI(
    model="models/gemini-2.5-flash",
    temperature=0.2
)

qa = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectordb.as_retriever(search_kwargs={"k": 5}),
    chain_type_kwargs={"prompt": PROMPT},
    return_source_documents=True
)

result = qa.invoke({
    "query": "Generate an MD&A draft for the most recent reporting period."
})
```

```python
md = "# Management’s Discussion and Analysis (Draft)\n\n"
md += result["result"]
md += "\n\n## Source Periods\n"

for doc in result["source_documents"]:
    md += f"- {doc.metadata['period']}\n"

print(md)

with open("MDNA_Draft.md", "w") as f:
    f.write(md)
```
